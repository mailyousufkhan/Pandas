```python
dataframe1=pd.DataFrame({"employee": ["ABC", "DEF", "XYZ"], "age": ["34","56","54"] })
dataframe1
```

```python
dataframe2=pd.DataFrame({"employee": ["ABC", "DEF", "HKJ"], "salary": ["34999","5676","54556"] })
dataframe2
```

```python
dataframe3=pd.merge(dataframe1,dataframe2, on="employee")
dataframe3
```

```python
dataframe3=pd.merge(dataframe1,dataframe2, on="employee", how="left")
dataframe3
```

```python
dataframe3=pd.merge(dataframe1,dataframe2, on="employee", how="right")
dataframe3
```

```python
dataframe3=pd.merge(dataframe1,dataframe2, on="employee", how="outer")
dataframe3
```

```python
fb_finance = pd.read_csv('C:/Users/Yousuf Khan/Downloads/files/FB.csv')
fb_finance.head()
```

```python
fb_finance['previous_closing_price']=fb_finance['Close'].shift(1)
fb_finance
```

```python
fb_finance['delta_closing_price']=fb_finance['Close'] - fb_finance['previous_closing_price']
fb_finance
```

```python
fb_finance['weekly_returns']=((fb_finance['Close'] - fb_finance['Close'].shift(7))/fb_finance['Close'])*100
fb_finance
```
